```python
class Shop:
    def __init__(self, name, max_scooters):
        self.name = name
        self.max_scooters = max_scooters
        self.available_scooters = max_scooters
        self.rentals = {}

    def rent_scooter(self, customer, scooters, duration, duration_type):
        if scooters > self.available_scooters:
            return "Not enough scooters available"
        self.available_scooters -= scooters
        rate = {'hour': 10, 'day': 20, 'week': 50}
        total_cost = rate[duration_type] * duration * scooters
        if 3 <= scooters <= 5:
            total_cost *= 0.85  # 15% discount
        self.rentals[customer.name] = total_cost
        return f"Scooter rented to {customer.name}. Total cost: {total_cost}"

    def return_scooter(self, customer_name):
        if customer_name not in self.rentals:
            return "No scooter rented by this customer"
        cost = self.rentals.pop(customer_name)
        return f"Total cost for {customer_name} is {cost}"

    def get_available_scooters(self):
        return self.available_scooters


class Customer:
    def __init__(self, name):
        self.name = name


def main():
    shop_name = input("Enter shop name: ")
    max_scooters = int(input("Enter maximum number of scooters for the shop: "))
    scooter_shop = Shop(shop_name, max_scooters)

    while True:
        print("\n1. Rent a Scooter\n2. Return a Scooter\n3. Check Available Scooters\n4. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            customer_name = input("Enter customer name: ")
            customer = Customer(customer_name)
            scooters = int(input("How many scooters to rent: "))
            duration = int(input("Enter duration of rent: "))
            duration_type = input("Enter duration type (hour/day/week): ")
            print(scooter_shop.rent_scooter(customer, scooters, duration, duration_type))

        elif choice == "2":
            customer_name = input("Enter customer name to return scooters: ")
            print(scooter_shop.return_scooter(customer_name))

        elif choice == "3":
            print(f"Available scooters: {scooter_shop.get_available_scooters()}")

        elif choice == "4":
            print("Exiting...")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
# Define a class for flights
class Flight:
    def __init__(self, number, origin, destination, departure, arrival, price):
        self.number = number # A string representing the flight number
        self.origin = origin # A string representing the origin city
        self.destination = destination # A string representing the destination city
        self.departure = departure # A string representing the departure time in HH:MM format
        self.arrival = arrival # A string representing the arrival time in HH:MM format
        self.price = price # A float representing the price of the flight in dollars
        self.available = True # A boolean indicating whether the flight is available or not

    def __str__(self):
        return f"Flight {self.number}: {self.origin} - {self.destination}, {self.departure} - {self.arrival}, ${self.price}"

# Define a class for reservations
class Reservation:
    def __init__(self, flight, customer):
        self.flight = flight # A Flight object representing the flight that is reserved
        self.customer = customer # A string representing the name of the customer
        self.confirmed = False # A boolean indicating whether the reservation is confirmed or not

    def __str__(self):
        return f"Reservation for {self.customer}: {self.flight}, Confirmed: {self.confirmed}"

# Define a list of flights
flights = [
    Flight("AA123", "New York", "Los Angeles", "08:00", "11:00", 300.0),
    Flight("BB456", "Chicago", "Houston", "09:30", "12:30", 200.0),
    Flight("CC789", "Miami", "San Francisco", "10:00", "13:00", 400.0)
]

# Define a list of reservations
reservations = []

# Define a function to display the available flights
def display_flights():
    print("The following flights are available:")
    for flight in flights:
        if flight.available:
            print(flight)

# Define a function to make a reservation
def make_reservation():
    display_flights() # Show the available flights
    number = input("Please enter the flight number you want to reserve: ") # Ask the user to enter the flight number
    flight = None # Initialize a variable to store the flight object
    for f in flights: # Loop through the flights list
        if f.number == number and f.available: # If the flight number matches and the flight is available
            flight = f # Assign the flight object to the variable
            break # Break the loop
    if flight is None: # If no flight is found
        print("Invalid or unavailable flight number. Please try again.") # Print an error message
        return # Return from the function
    customer = input("Please enter your name: ") # Ask the user to enter their name
    reservation = Reservation(flight, customer) # Create a reservation object
    reservations.append(reservation) # Append the reservation object to the reservations list
    print("Your reservation has been made. Please confirm it within 24 hours.") # Print a confirmation message
    print(reservation) # Print the reservation details

# Define a function to confirm a reservation
def confirm_reservation():
    customer = input("Please enter your name: ") # Ask the user to enter their name
    reservation = None # Initialize a variable to store the reservation object
    for r in reservations: # Loop through the reservations list
        if r.customer == customer and not r.confirmed: # If the customer name matches and the reservation is not confirmed
            reservation = r # Assign the reservation object to the variable
            break # Break the loop
    if reservation is None: # If no reservation is found
        print("No reservation found for your name. Please try again.") # Print an error message
        return # Return from the function
    reservation.confirmed = True # Set the confirmed attribute to True
    reservation.flight.available = False # Set the available attribute of the flight to False
    print("Your reservation has been confirmed. Thank you for choosing us.") # Print a confirmation message
    print(reservation) # Print the reservation details

# Define a function to cancel a reservation
def cancel_reservation():
    customer = input("Please enter your name: ") # Ask the user to enter their name
    reservation = None # Initialize a variable to store the reservation object
    for r in reservations: # Loop through the reservations list
        if r.customer == customer: # If the customer name matches
            reservation = r # Assign the reservation object to the variable
            break # Break the loop
    if reservation is None: # If no reservation is found
        print("No reservation found for your name. Please try again.") # Print an error message
        return # Return from the function
    reservations.remove(reservation) # Remove the reservation object from the reservations list
    reservation.flight.available = True # Set the available attribute of the flight to True
    print("Your reservation has been cancelled. We hope to see you again.") # Print a cancellation message
    print(reservation) # Print the reservation details

# Define a function to display the menu
def display_menu():
    print("Welcome to the airline reservation system. Please choose an option:")
    print("1. Make a reservation")
    print("2. Confirm a reservation")
    print("3. Cancel a reservation")
    print("4. Exit")

# Define a main function
def main():
    while True: # Loop indefinitely
        display_menu() # Display the menu
        choice = input("Please enter your choice: ") # Ask the user to enter their choice
        if choice == "1": # If the choice is 1
            make_reservation() # Call the make_reservation function
        elif choice == "2": # If the choice is 2
            confirm_reservation() # Call the confirm_reservation function
        elif choice == "3": # If the choice is 3
            cancel_reservation() # Call the cancel_reservation function
        elif choice == "4": # If the choice is 4
            print("Thank you for using the airline reservation system. Goodbye.") # Print a farewell message
            break # Break the loop
        else: # If the choice is invalid
            print("Invalid choice. Please try again.") # Print an error message

# Call the main function
if __name__ == "__main__":
    main()
```
